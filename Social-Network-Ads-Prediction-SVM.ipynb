### Spliting the dataSet  into train ,test and validate dataSet



line 1: To Check the structure and information of the Dataset.



```python
# To check the correct DataFrame
print("DataSet information")
print(dataSet.info())
print("DataSet columns")
print(dataSet.columns)

dataSet
```

#As we see Gender column is in string format so we have to convert it to integer

```python
#convert the Gender column to numaric value to be able to classifier 
convert = {"Gender": {"Female": 1, "Male": 0}}#convert "Female" to 1 and "Male" to 0
dataSet = dataSet.replace(convert)#replace the values in the dataset
dataSet #display the dataSet after change
```

line 1: import train_test_split to split the dataSet
line 2,3: seperate the data to input and output 
line 7: spliting the dataSet  into train ,test dataSet


```python
from sklearn.model_selection import train_test_split
#Seperate the dataSet to input and target 
inputAtt = dataSet.values[:, 1:3]
target = dataSet.values[:, 4]

#Split the dataSet into train set and remain (test and validate) set
inputAtt_train, inputAtt_remain, target_train, target_remain = train_test_split(inputAtt, target, train_size=0.8,random_state= 100)
#Split the remain set into test and validate sets
inputAtt_valid, inputAtt_test, target_valid, target_test = train_test_split(inputAtt_remain, target_remain, test_size=0.5,random_state= 100)
```

### Setting up the selected ML model and with the default or set patameters

line 1: import SVC from sklearn.svm to create the svm classifier
line 3: create the classifier
line 5,6: print the default parameters of the SVM classifier

```python
from sklearn.svm import SVC #for Support Vector Classifier

classifier = SVC(kernel='linear', random_state=0,probability=True)  # initialize SVM classifier
# Print the default parameters of the SVM classifier
print("Default parameters of SVM classifier:")
print(classifier.get_params())

```

### Training the ML model with dataSet

```python
# Fit the classifier and give it labeled data (input + output) to teach it.
classifier.fit(inputAtt_train, target_train)

# Print the size of train and test data 
print("Size of training data set:", inputAtt_train.shape[0])
print("Size of testing data set:", inputAtt_test.shape[0])
print("Size of validate data set:", inputAtt_valid.shape[0])
```

### Testing and evaluating the trained ML model(s) against the test dataset and report the result visually

```python
from sklearn.metrics import accuracy_score, confusion_matrix #Import functions for evaluating classification performance
# To make predictions on the testing data
target_pred = classifier.predict(inputAtt_test)
# To obtain the predicted probabilities of the classes for each instance in the test set
target_prob = classifier.predict_proba(inputAtt_test)

#Calculate and print the accuracy of the model predictions
accuracy = accuracy_score(target_test, target_pred)
print("Accuracy:", accuracy)
```

### Result:

The accuracy of the classifier is 0.8

# visualizing the result:

#Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, classification_report#import functions for classification report
#To Create a confusion matrix
conf_matrix = confusion_matrix(target_test, target_pred)

# Print confusion matrix
confusion_matrix = confusion_matrix(target_test, target_pred)
print(confusion_matrix)

plt.figure(figsize=(6,6))
sns.heatmap(confusion_matrix,annot = True)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
# Print classification report
print(classification_report(target_test, target_pred))
```

#ROC Curve and AUC

Receiver Operating Characteristic (ROC) curve is a graphical representation of a classifier's performance across different thresholds.

line 1: import roc_curve and auc from sklearn.metrics to visualizing the result
line 3: function to computes ROC curve points. give it the target test and the predicted probabilities and it return 3 values:
- fpr: False Positive Rate for each threshold.
- tpr: True Positive Rate for each threshold.
- _: Thresholds used to compute fpr and tpr.
line 4: function to computes the Area Under the ROC Curve (AUC), and AUC is single scalar value summarizing the performance of the classifier.
line 5: size of the figure
line 6: plots the curve and give it the returned values(fpr,tpr) and specifies the color and the width of the curve, and add lable for the AUC value (formatted to two decimal )
line 7: plots a diagonal line from (0,0) to (1,1) that represent a classifier with random guessing. and specifies the color and the width and the style of the line 
line 8,9: the range of x-axis and y-axis
line 10,11: lables for x-axis and y-axis
line 12: title of the plot
line 13: specifies the place for the lable created in line 6
line 14: display the plot

```python
from sklearn.metrics import roc_curve,auc 

fpr, tpr, _ = roc_curve(target_test, target_prob[:, 1], pos_label=1)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(6,4))
plt.plot(fpr, tpr, color='green', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
